### concat()

```python
#merging the series
x = pd.Series(['Reem','Layla'])
y = pd.Series(['Mona','Dina'])

z=pd.concat([x,y]) # this function merge both
z
```

```python
x = pd.Series(['Reem','Layla'])
y = pd.Series(['Mona','Dina'])

z=pd.concat([x,y], ignore_index = True) # this function merge both & change the index from 0 to 3
z
```
