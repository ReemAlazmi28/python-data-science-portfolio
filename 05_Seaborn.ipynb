# Histogram

```python
#sns.histplot(df['mpg']);
plt.figure(figsize=(10,5));
sns.histplot(df, x = 'mpg',bins=20, hue="origin", multiple='stack', alpha =0.4); # alpha to the calrety of the color and it btween 0 and 1
plt.title("MPG distrbution");
#hue is to add cocrs
#multiple=
        # layer
        # fill
        # dodge
```

# Jointplot

```python
sns.jointplot(x='mpg', y = 'weight', data =df, kind ='scatter' );# the relayionship btween two varibles
#kind =
#hex
#kde
#reg
```

# PairPlot

```python
sns.pairplot(df, hue='origin', palette = 'coolwarm')
plt.show()
```

```python
# to change the color we change the palette = 'the color pallet we want'
```

# KdePlot

```python
sns.kdeplot(data=df , x= 'mpg', hue ='origin' );
plt.title('KDE Chart');
```

```python
sns.kdeplot(df , x= 'mpg' );
plt.title('KDE Chart'); # the geniral dinsty of the data
```

# BarPlot

# CountPlot

```python
sns.countplot(x = 'origin',data = df); # only takes one variable in this case its the x
```

```python
# creat a data frame and a new column that apply a spesific function taht takes a  condition
df['mpg_level']=df['mpg'].apply(lambda x: 'low' if x <15 else 'High' if x>30 else 'Medium')
df.head()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x ='model_year', hue= 'mpg_level',data=df);
```

```python
sns.countplot(x ='model_year', hue= 'cylinders',data=df);
```

# RelPlot

```python
sns.relplot(x ='model_year', y= 'mpg',data=df); # Direct relation
```

```python
sns.relplot(x ='horsepower', y= 'mpg',hue='origin', size= 'weight',data=df); #inverse relation
```

# CatPlot

```python
sns.catplot(data = df , kind='bar' ,x= 'origin', y='horsepower', hue= 'cylinders', alpha=0.5,height=4);
```
