# Univariate Analysis
 numerical :
- histogram
- density plot
  
 categorical :
- bar plot
- pie plot

```python
sns.histplot(data=diamonds,x='price',bins=10); # one variable only
```

```python
sns.histplot(data=tips,x='total_bill',bins=10);
```

```python
sns.kdeplot(data=tips,x='total_bill');
```

```python
sns.displot(data=tips,x='total_bill',kind= 'hist', col='sex');
# i can change the kind to histogram or kde by : kind= 'hist'
# in displo we use : col for compearing
```

```python
sns.displot(data=tips,x='total_bill', kde=True);
```

```python
# categorical data
sns.countplot(data = diamonds,x='cut');
#sns.countplot(data = diamonds,x='color');
##sns.countplot(data = diamonds,x='clarity');
```

```python
plt.pie(tips.smoker.value_counts(), labels=['Non-Smoker','Smoker'],autopct="%1.2f%%");
```

```python
# to be in the safe side use this
plt.pie(tips.smoker.value_counts(), labels=tips.smoker.value_counts().index,autopct="%1.2f%%");
```

# Categorical Vs. Categorical

```python
sns.barplot(data = tips, x='day', y ='total_bill', hue='sex');
```

```python
sns.histplot(diamonds.sort_values('clarity').reset_index(), x='clarity', hue='cut',alpha=0.6, multiple="dodge");
```

# Numerical Vs. Categorical

```python
sns.barplot(data=diamonds,x='color',y='price');
```

```python
sns.swarmplot(data=tips, x='time',y='tip');
```
