#### How many move that got released in each year

```python
movies_count = df.release_year.value_counts(ascending=True)
movies_count
```

```python
movies_count = movies_count.reset_index()
movies_count
```

```python
movies_count.columns =['year','count']
movies_count = movies_count.sort_values(by=['year'])
movies_count
```

```python
plt.plot(movies_count['year'],movies_count['count'] , c='0.2');
```

```python
plt.plot(movies_count['year'],movies_count['count'] , c='0.2');
plt.xlabel("Years");
plt.ylabel("Movie Count");
plt.title("Number of released Movies");
plt.xlim(1960,2015); # choose the limit of the axis
plt.xticks(range(1960,2016,5)); # the gape is 5 years from1960 to 2015
# plt.axis('off') # to dele the axis x and y
```

```python
plt.plot(movies_count['year'],movies_count['count'] , c ='red',marker = 'o',markersize =10 ,markerfacecolor='pink',
        markeredgecolor= 'gray', markeredgewidth=4);
plt.plot(range(1960,2015,10),[500,400,300,200,100,600], c= 'green', linestyle = ':', linewidth=3);
plt.legend(['Moves','Series'], shadow=True, loc='center') # loc = (x,y)
plt.xlabel("Years");
plt.ylabel("Movie Count");
plt.title("Number of released Movies");
plt.xlim(1960,2015); # choose the limit of the axis
plt.xticks(range(1960,2016,5)); # the gape is 5 years from1960 to 2015
# plt.axis('off') # to dele the axis x and y
```

### plot the 4 top revenue

```python
top_revenue = df.groupby('original_title').sum().sort_values(by='revenue', ascending= False)
top_revenue.head()
```

```python
movies = top_revenue.head().index
movies # only the titel
```

```python
revenue = top_revenue.head().revenue
revenue # title and revenue
```

```python
movies = top_revenue.head().index
revenue = top_revenue.head().revenue.values
revenue # only revenue
```

```python
movies = ['Avatar','Star Wars' ,'Titanic','The Avengers','JW']
revenue = top_revenue.head().revenue.values
```

```python
plt.figure(figsize = (12,10))
color= ['blue','#1147B2']
plt.bar(movies, revenue,color = color);
plt.xlabel('movies',fontsize=20);
plt.ylabel('revenue',fontsize=20);
plt.title('Top 5 Movies Bssed on Revenue');
```

```python
top_genres = df.groupby('genres').sum().sort_values(by='revenue', ascending=False)['revenue'][0:5]
top_genres
```

# Scatter Chart

```python
plt.figure(figsize = (2,4));
plt.scatter(np.random.randn(500),np.random.randn(500));
```

```python
plt.scatter(df.vote_average,df.revenue, alpha= 0.7, label =' vote average');
plt.scatter(df.vote_count,df.revenue, alpha= 0.7 , marker= '+' ,s = 200 , label='vote count');
plt.legend(loc=1);
plt.xlabel('vote average & count ');
plt.ylabel('revenue')
plt.title('Relatioship between vote average and revenue & count vs. Revenue')
```
